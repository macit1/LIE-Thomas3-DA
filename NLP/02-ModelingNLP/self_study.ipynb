```python
import spacy
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# İngilizce dil modelini yükleyin
nlp = spacy.load("en_core_web_sm")

# İki cümleyi vektörize et
sentence1 = "I love learning"
sentence2 = "I learn loving"
doc1 = nlp(sentence1)
doc2 = nlp(sentence2)

# Her bir kelimenin vektörlerini al
word_vectors1 = [token.vector for token in doc1]
word_vectors2 = [token.vector for token in doc2]

# Kelime vektörlerini birleştirin
all_word_vectors = np.vstack([word_vectors1, word_vectors2])

# t-SNE boyut indirgeme uygula
tsne = TSNE(n_components=2, perplexity=5, random_state=42)
word_vectors_tsne = tsne.fit_transform(all_word_vectors)

# Görselleştirmek için her kelimenin konumunu alın
x = word_vectors_tsne[:, 0]
y = word_vectors_tsne[:, 1]

# Her bir kelimenin etiketini alın
labels = [token.text for token in doc1] + [token.text for token in doc2]

# Kelimeleri görselleştirin
plt.figure(figsize=(10, 6))
plt.scatter(x, y, marker='o')

# Her kelimenin etiketini ekle
for label, xi, yi in zip(labels, x, y):
    plt.annotate(label, xy=(xi, yi), xytext=(0, 0), textcoords='offset points')

plt.show()
```
