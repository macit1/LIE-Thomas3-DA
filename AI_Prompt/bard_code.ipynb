```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained GPT-2 model and tokenizer
model = GPT2LMHeadModel.from_pretrained("EleutherAI/gpt-neo-2.7B")
tokenizer = GPT2Tokenizer.from_pretrained("EleutherAI/gpt-neo-2.7B")

# Set up your input prompt
prompt = "Your chatbot input prompt goes here."

# Tokenize the prompt
input_ids = tokenizer.encode(prompt, return_tensors="pt")

# Generate a response using the model
response_ids = model.generate(input_ids, max_length=150, num_beams=5, no_repeat_ngram_size=2, top_k=50, top_p=0.95, temperature=0.7)

# Decode the generated response
response_text = tokenizer.decode(response_ids[0], skip_special_tokens=True)

# Print the generated response
print("Input Prompt:")
print(prompt)
print("\nGenerated Response:")
print(response_text)
```
